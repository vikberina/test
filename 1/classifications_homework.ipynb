```python
from sklearn.datasets import load_iris

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame(load_iris().data, columns = load_iris().feature_names)
iris = load_iris()
X = df.drop(['petal length (cm)', 'petal width (cm)'], axis = 1)
Y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)
print(f'предсказание: {y_pred}') 
```

```python
ax=sns.scatterplot(x=X_test['sepal length (cm)'], y=X_test['sepal width (cm)'], palette='Set1',hue=y_pred)
ax=sns.scatterplot(x=lda.means_[:,0], y=lda.means_[:,1], s=400, marker='*', color='orange')
ax.legend(loc = 'lower right')
ax.set_title('Предсказание для X_test')
```

```python
kms = KMeans(n_clusters=3)
clusters = kms.fit_predict(X)
```

```python
ax=sns.scatterplot(x=X['sepal length (cm)'], y=X['sepal width (cm)'], palette='Set1',hue=clusters)
ax.set_title('Кластеризация для X')
```
